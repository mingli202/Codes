```python
import json

with open("out.json", "r") as file:
    arr = json.loads(file.read())

professors = set()


for i in arr:
    professors.add(i["lecture"]["prof"])

    if "prof" in i["lab"]:
        professors.add(i["lab"]["prof"])

professors.remove("")

with open("professors.json", "w") as file:
    file.write(json.dumps(list(professors)))
```
